#groupby question
#who has highest average store?

```python
import random
random.seed(444)
a = [random.choice(['NY','CH','TX']) for i in range(1000)]
x = [random.randint(1,1100)for i in range(1000)]
df = pd.DataFrame({'STATE': a,"score": x})
df
```

```python
new_df = df.groupby('STATE').agg({'score':'mean'})
new_df
```

#barchart

```python
fig,ax = plt.subplots()
ax.bar(new_df.index,new_df.score)
```

#pie chart

```python
fig,ax = plt.subplots()
ax.pie(new_df.score,labels = new_df.index);
```
